```python
import pandas as pd
import os 
import numpy as np
import folium
from geopy.geocoders import Nominatim
from folium import FeatureGroup, LayerControl, Map, Marker
from folium.plugins import HeatMap
import re
```

```python
data_1997 = pd.read_csv("1997_completo.csv", index_col= 'Unnamed: 0')
data_2017 = pd.read_csv("2017_completo.csv", index_col= 'Unnamed: 0')
```

```python
data_1997.shape, data_2017.shape
```

```python
def coords(df):
    lat = []
    lon = []
    for i in range(len(df)):
        ex1 = str(df['Lat_Lon'].iloc[i])
        match1 = re.match(r"^.*\((.*)\,.*$",ex1)
        match2 = re.match(r"^.*\,(.*)\).*$",ex1)
        lat_prov = float(match1.group(1))
        lon_prov = float(match2.group(1))
        lat.append(lat_prov)
        lon.append(lon_prov)
    lat_lon = list(zip(lat, lon))
    return lat_lon
```

```python
coords_1997 = coords(data_1997)
coords_2017 = coords(data_2017)
```

```python
### para 1997
def plot_map(coords):
    rand_loc = np.arange(0,len(coords))
    np.random.shuffle(rand_loc)
    rand_loc
    m = folium.Map(coords[rand_loc[0]])
    HeatMap(coords).add_to(m)
    return m
```

```python
plot_map(coords_1997)
```

```python
plot_map(coords_2017)
```
