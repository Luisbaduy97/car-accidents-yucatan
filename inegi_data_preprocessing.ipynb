```python
import pandas as pd
import os 
import numpy as np
import folium
from geopy.geocoders import Nominatim
from folium import FeatureGroup, LayerControl, Map, Marker
from folium.plugins import HeatMap
```

```python
lista_estados = pd.read_csv("../tabla_completa_estados_municipios.csv", engine = 'python')
lista_estados = lista_estados.sort_values(by = ['Nom_Ent', 'Cve_Mun'])
clave = lista_estados.Cve_Ent.unique()
clave = clave.tolist()
entidad = lista_estados.Nom_Ent.unique()
entidad = entidad.tolist()
dict_estados = {}
for i in range(0, len(clave)):
    dict_estados.setdefault(clave[i], entidad[i])
yuc = lista_estados[lista_estados["Nom_Ent"] == 'Yucatan']
mainpath = "D:/Datasets_Juego/Accidentes_Historicos/conjunto_de_datos"
datos_inegi = pd.read_csv(os.path.join(mainpath + "/" + "atus_anual_2017.csv"))
lista_entidades = []
for j in range(0,len(datos_inegi)):
    lista_entidades.append(dict_estados[datos_inegi.ID_ENTIDAD[j]])
datos_inegi["ID_ENTIDAD"] = lista_entidades
yucatan = datos_inegi[datos_inegi["ID_ENTIDAD"] == 'Yucatan']
#######Vamos a agregar los municipios#####
numero_mun = yuc.Cve_Mun.unique()
numero_mun = numero_mun.tolist()
municipio = yuc.Nom_Mun.unique()
municipio = municipio.tolist()
dir_municipios = {}
lista_municipios = []
for h in range(len(yuc)):
    dir_municipios.setdefault(numero_mun[h], municipio[h])
for z in range(0, len(yucatan)):
    lista_municipios.append(dir_municipios[yucatan.ID_MUNICIPIO.iloc[z]])
yucatan["ID_MUNICIPIO"] = lista_municipios
####Hasta aqui el dataset ya tiene agregado el estado y sus municipios####
```

Vamos a ponerle latitud y longitud

```python
yucatan["CITY_STATE"] = list(yucatan.ID_MUNICIPIO + ', ' + yucatan.ID_ENTIDAD)
```

```python
ubicaciones = yucatan.CITY_STATE.unique()
latitude = []
longitude = []
geolocation = Nominatim(timeout=5)
for p in range(len(ubicaciones)):
    location = geolocation.geocode(ubicaciones[p])
    latitude.append(location.latitude)
    longitude.append(location.longitude)
```

```python
puntos = list(zip(latitude, longitude))
```

```python
localizaciones_estados = {}
for k in range(len(ubicaciones)):
    localizaciones_estados.setdefault(ubicaciones[k], puntos[k])
```

```python
posicion_final = []
for t in range(0,len(yucatan)):
    posicion_final.append(localizaciones_estados[yucatan.CITY_STATE.iloc[t]])
```

```python
yucatan["Lat_Lon"] = posicion_final
```
